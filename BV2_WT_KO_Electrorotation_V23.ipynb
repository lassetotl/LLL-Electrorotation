```python
fig = plt.figure(figsize = (10, 8))
plt.title('BV-2 WT & KO rotation spectra', fontsize = 18)
plt.xlabel('Signal frequency [MHz]', fontsize = 18)
plt.ylabel('Cell rotation [rpm]', fontsize = 18)
plt.xscale('symlog')

a = Rotation_Spectra(ERM_WT, freq_WT, 'darkslategray', 'BV-2 WT', detail = 0)
b = Rotation_Spectra(ERM_KO, freq_, 'chocolate', 'BV-2 KO 24.03.23')
c = Rotation_Spectra(ERM_KO_, freq_, 'sandybrown', 'BV-2 KO 29.03.23')
d = Rotation_Spectra(ERM_KO__, freq_, 'darkorange', 'BV-2 KO 14.04.23')
e = Rotation_Spectra(ERM_KO___, freq_, 'orange', 'BV-2 KO 14.04.23')

plt.axhline(0, color = 'white', lw = 3.5, zorder=-100)
plt.xlim(xmin = -0.2); plt.ylim(ymin = -13)
plt.grid(1)#; plt.legend(prop={'size': 13})
#plt.savefig('BV2_KO_Rotation')
plt.show()
```
